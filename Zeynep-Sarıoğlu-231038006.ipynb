```python
#Temel Calisan Sinifi Olusturma
class Calisan:
    def bilgi_goster(self):
        return f" Ad: {self.ad}, Yas: {self.yas}, Pozisyon: {self.pozisyon}, Maas: {self.maas}"
    

#Ofis Calisani Sinifi Olusturma
class OfisCalisani(Calisan):
    def bilgi_goster(self):
        return super().bilgi_goster() + f" , Calisma Saati:{self.calisma_saati} saat"
    

#SahaCalisani Sinifi Olusturma
class SahaCalisani(Calisan):
    def bilgi_goster(self):
        return super().bilgi_goster() + f" , Saha Bolgesi:{self.saha_bolgesi} bolge"
    
#Calisan Listesi Olusturma
calisanlar = []

#Calisan Ekleme Fonksiyonu
def calisan_ekle(tur, ad, yas, pozisyon, maas, ekstra):
    if tur == "ofis":
        calisan = OfisCalisani()
        calisan.calisma_saati = ekstra
    elif tur == "saha":
        calisan = SahaCalisani()
        calisan.saha_bolgesi = ekstra
    else:
        calisan = Calisan()

        calisan.ad = ad
        calisan.yas = yas
        calisan.pozisyon = pozisyon
        calisan.maas = maas
        calisanlar.append(calisan)


#Calisan Listeleme Fonksiyonu
def calisanlari_listele():
    for calisan in calisanlar:
        print(calisan.bilgi_goster())


#Test Etme
if __name__ == "__main__":
    calisan_ekle("Ofis" , "Zeynep" , "20" , "IT Uzmani" , 40000 , 8)
    calisan_ekle("Saha" , "Miray" , "25" , "Satis Danismani" , 20000 , "Marmara")
    calisan_ekle("Temel" , "Emirhan" , "20" , "Temsilci" , 30000 , None)

calisanlari_listele()





 
    
```
